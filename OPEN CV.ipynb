```python
import cv2
```

```python
#DISPLAYING AND READING IMAGE
fig=cv2.imread("C:\\Users\\nqnqa\\Downloads\\grass-pictures-dkxrgx8eyd7ua8be.webp")
cv2.imshow("img",fig)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#SHAPE

fig.shape
```

```python
#MATPLOTLIB
import matplotlib.pyplot as plt
plt.imshow(fig)
```

```python
#COLOR CONVERT
fig_01=cv2.cvtColor(fig,cv2.COLOR_RGB2BGR)

plt.imshow(fig_01)         #  >>  "MATPLOTLIB READS BGR"
```

```python
# RGB TO GRAY 
fig_03=cv2.cvtColor(fig,cv2.COLOR_RGB2GRAY)

cv2.imshow("img",fig_03)

cv2.waitKey(0)

cv2.destroyAllWindows()
```

```python
#EDGE DETECTION
fig_05=cv2.cvtColor(fig,cv2.COLOR_BGR2GRAY)     # EDGE DETECTION WORKS MAINLY FOR GRAY 
sobel_xy=cv2.Sobel(fig,ddepth=cv2.CV_64F,dx=1,dy=1)
cv2.imshow("img",sobel_xy)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
